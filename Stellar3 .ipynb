## 第三次恒星物理编程作业

原子核$i,j$之间克服库伦势垒进行热核聚变的反应截面S：
$$
S\propto exp\left(-\frac{\pi Z_i Z_j e^2}{\varepsilon_0 hv}\right)exp\left(-\frac{m_g v^2}{2kT}\right)
$$

$$
m_g=\frac{m_{g,i}m_{g,j}}{m_{g,i}+m_{g,j}}
$$

#### Gamow peak $\frac{dS}{dv}=0$:
$$
v_p=\left(\frac{\pi Z_i Z_j e^2 k T}{\varepsilon_0 h m_g} \right)^{1/3}
$$

计算时取，$\frac{1}{4\pi\varepsilon_{0}}=1,h = 6.626 × 10^{−27}erg\cdot s， k = 1.381 × 10^{−16}erg\cdot K^{-1}, Z_i=Z_j=1, m_{g,i}=m_{g,j}=m_p=1.67\times 10^{-24} g, T=1.0\times 10^{7}K $

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.constants as sc

fig=plt.figure(figsize=(20,8))
ax1 = plt.subplot(131)
ax2=plt.subplot(132)
ax3 = plt.subplot(133)

v = np.linspace(1, 5*10**6, 100)
e=math.e
Q_T=e**(- (sc.pi*sc.e**2)/(sc.epsilon_0*sc.h*v))
velo=e**(-(sc.m_p/2 * v**2)/(2*sc.k*(10**7)))
S=e**(-(sc.pi*sc.e**2)/(sc.epsilon_0*sc.h*v))*e**(-(sc.m_p/2 * v**2)/(2*sc.k*(10**7)))
#gravitational potential plot 
ax1.plot(v ,S, lw=1,c="b")
ax1.grid(True)
ax1.set_ylabel(r"$S$",fontsize=20)
ax1.set_xlabel(r"$v$",fontsize=20)
ax1.set_title(r"S - v",fontsize=20)

#量子隧道贯穿几率
ax2.plot(v,Q_T, lw=1, c="g")
ax2.grid(True)
ax2.set_ylabel(r"penetration probability",fontsize=20)
ax2.set_xlabel(r"v",fontsize=20)
ax2.set_title(r"Penetration probability - v",fontsize=20)

#Maxwell velocity distribution
ax3.plot(v,velo, lw=1, c="y")
ax3.grid(True)
ax3.set_ylabel(r"velocity distribution",fontsize=20)
ax3.set_xlabel(r"v",fontsize=20)
ax3.set_title("velocity distribution - v",fontsize=20)
```

从上面左图中可以读出Gamow Peak出现在速率约为$1.3\times 10^6$处，利用前面给出的$v_p$公式，计算得：

```python
((sc.pi*sc.k*10**7*(sc.e**2))/(sc.epsilon_0*sc.h*sc.m_p/2))**(1/3)
```

即相对误差：
$$
\varepsilon=\frac{|1.3\times 10^6-v_p|}{v_p}\times 100\%=1.084\%
$$
