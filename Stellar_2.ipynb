## 恒星物理第二次编程作业
首先计算相关的物理量：
$$
\rho(r)=\rho_c [1-(\frac{r}{R})^2]
$$
$$
m(r)=\int_{0}^{r}4\pi r^2 \rho(r) dr=4\pi \rho_c (\frac{r^3}{3}-\frac{r^5}{5R^2})
$$
利用边界条件$m(R)=M$,可得：
$$
M=\frac{8\pi \rho_c R^3}{15} \Longrightarrow \rho_c=\frac{15M}{8\pi R^3}
$$

### （1）.
$$
g(r)=\frac{Gm(r)}{r^2}=4\pi G \rho_c(\frac{r}{3}-\frac{r^3}{5R^2})
$$

$$
\frac{g(r)}{GM/R^2}=\frac{15}{2}(\frac{r}{3R}-\frac{r^3}{5R^3})
$$

```python
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(20,8))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)


r = np.linspace(0.001, 1, 200)

#gravitational potential plot 
g_r=15/2*(r/3-r**3/5)
ax1.plot(r,g_r, lw=1, label="g_r_1",c="b")
ax1.grid(True)
ax1.set_xlim(0,1)
ax1.set_ylim(0,2)
ax1.set_ylabel(r"$\frac{g(r)}{GM/R^2}$",fontsize=20)
ax1.set_xlabel(r"$\frac{r}{R}$",fontsize=20)
ax1.set_title(r"relation between $\frac{g(r)}{GM/R^2}$ and $\frac{r}{R}$",fontsize=20)

#双对数坐标轴
ax2.plot(r,g_r, lw=1, label="g_r_2",c="b")
ax2.grid(True)
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_ylabel(r"$log(\frac{g(r)}{GM/R^2}$)",fontsize=20)
ax2.set_xlabel(r"$log(\frac{r}{R})$",fontsize=20)
ax2.set_title(r"$log(\frac{g(r)}{GM/R^2}$) - log($\frac{r}{R}$)",fontsize=20)
```

### (2).
$$
\frac{dP}{dr}=-\rho(r)g(r)
$$

$$
P(R)-P(r)=-\int_r^R \rho(r)g(r) dr \Longrightarrow P(r)=G\int_r^R\rho\frac{m}{r^2} dr
$$

$$
P(r)=4\pi G\rho_c^2 \int_r^R[1-(\frac{r}{R})^2](\frac{r}{3}-\frac{r^3}{5R^2}) dr
$$

若记$t=r/R$,并且引入$P_c=\frac{15GM^2}{16\pi R^4}$,上述积分得结果可表达为：
$$
\frac{P(r)}{P_c}=15(\frac{1-t^2}{6}-\frac{2-2t^4}{15}+\frac{1-t^6}{30})
$$

即：
$$
\frac{P(r)}{P_c}=1-\frac{15t^2}{6}+2t^4-\frac{t^6}{2}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(20,8))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)


r = np.linspace(0.001, 1, 200)

#gravitational potential plot 
p_r=1-15*r**2/6+2*r**4-r**6/2
ax1.plot(r,p_r, lw=1, label="p_r_1",c="g")
ax1.grid(True)
ax1.set_xlim(0,1)
ax1.set_ylim(0,1)
ax1.set_ylabel(r"$\frac{P(r)}{P_c}$",fontsize=20)
ax1.set_xlabel(r"$\frac{r}{R}$",fontsize=20)
ax1.set_title(r"relation between $\frac{P(r)}{P_c}$ and $\frac{r}{R}$",fontsize=20)

#双对数坐标轴
ax2.plot(r,p_r, lw=1, label="p_r_2",c="g")
ax2.grid(True)
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_ylabel(r"$log(\frac{P(r)}{P_c}$)",fontsize=20)
ax2.set_xlabel(r"$log(\frac{r}{R})$",fontsize=20)
ax2.set_title(r"$log(\frac{P(r)}{P_c}$) - log($\frac{r}{R}$)",fontsize=20)
```
